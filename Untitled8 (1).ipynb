```python
def get_embedding(text):
    response = client.embeddings.create(
        model="text-embedding-3-small",
        input=text
    )
    return response.data[0].embedding
```

```python
def cosine_similarity(a, b):
    dot_product = sum(x * y for x, y in zip(a, b))
    norm_a = sum(x**2 for x in a) ** 0.5
    norm_b = sum(x**2 for x in b) ** 0.5
    if norm_a == 0 or norm_b == 0:
        return 0
    return dot_product / (norm_a * norm_b)
```

```python
# This will take a few seconds and cost a fraction of a cent
block_embeddings = []

for block in blocks:
    embedding = get_embedding(block)
    block_embeddings.append(embedding)
    print(".", end="")  # progress dot

print(f"\nEmbedded {len(block_embeddings)} blocks.")
```

```python
def find_best_match_semantic(query):
    query_embedding = get_embedding(query)
    
    best_block = None
    best_score = -1
    
    for i, block_embedding in enumerate(block_embeddings):
        score = cosine_similarity(query_embedding, block_embedding)
        if score > best_score:
            best_score = score
            best_block = blocks[i]
    
    return best_block, best_score
```

```python
test_queries = [
    "maternity leave",
    "what to wear to work",
    "how to schedule a room",
    "retirement savings",
    "how many sick days do I get"
]

for query in test_queries:
    best_block, score = find_best_match_semantic(query)
    print(f"Query: {query}")
    print(f"Result: {best_block}")
    print(f"Score: {score:.2f}")
    print("-" * 50)
```
